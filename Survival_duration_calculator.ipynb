```python
def Survival_duration_calculator(age, time_unit):
    # Conversion factors
    months_in_year = 12
    weeks_in_year = 52.1775  # accounting for leap years
    days_in_year = 365.25    # accounting for leap years
    hours_in_day = 24
    minutes_in_hour = 60
    seconds_in_minute = 60

    # Standardize the time_unit input
    time_unit = time_unit.lower()

    # Map single letters to full time unit names
    time_unit_mapping = {
        'm': 'months',
        'w': 'weeks',
        'd': 'days',
        'h': 'hours',
        'mi': 'minutes',
        's': 'seconds'
    }

    # Get the full name of the time unit
    if time_unit in time_unit_mapping:
        time_unit = time_unit_mapping[time_unit]
    elif time_unit in time_unit_mapping.values():
        # Ensure that valid full names are also accepted
        time_unit = time_unit
    else:
        raise ValueError("Invalid time unit. Please choose from Months, Weeks, Days, Hours, Minutes, Seconds (or their first letter).")

    # Calculations based on the time unit
    if time_unit == 'months':
        duration = age * months_in_year
    elif time_unit == 'weeks':
        duration = age * weeks_in_year
    elif time_unit == 'days':
        duration = age * days_in_year
    elif time_unit == 'hours':
        duration = age * days_in_year * hours_in_day
    elif time_unit == 'minutes':
        duration = age * days_in_year * hours_in_day * minutes_in_hour
    elif time_unit == 'seconds':
        duration = age * days_in_year * hours_in_day * minutes_in_hour * seconds_in_minute

    return duration

def main():
    try:
        # Input from the user
        age = float(input("What's your age? "))
        time_unit = input("Please choose time unit: Months (M), Weeks (W), Days (D), Hours (H), Minutes (Mi), Seconds (S). ").strip().capitalize()

        # Normalize time unit for comparison
        if time_unit == 'Mi':
            time_unit = 'mi'
        else:
            time_unit = time_unit[0]

        # Calculate the duration
        duration = Survival_duration_calculator(age, time_unit)

        # Display the result
        print(f"The duration of {age} years in {time_unit} is: {duration:.2f}")
    except ValueError as e:
        print(e)

if __name__ == "__main__":
    main()
```
