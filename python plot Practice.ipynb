```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
plt.figure(figsize=(6,6))
plt.bar(["jan","feb","mar","Apr"],[22,33,41.5,44],width=0.3,label="mark1",color=["gold","black","olive","green"],log=True)
plt.legend()
plt.title("Bar Plot",color='red',size=11)
plt.xticks(rotation=45)
```

```python
month=["jan","feb","mar","Apr","may","jun","jul","aug","sep","oct","nov","dec"]
x=np.arange(0,len(month))-0.14
x_=np.arange(0,len(month))+0.25
x1=[18,23,33,43,32,56,76,88,91,78,51,29]
x2=[19,21,36,47,55,76,43,82,101,80,57,33]
plt.bar(x,x1,width=0.4,label="x1")
plt.bar(x_,x2,width=0.4,label="x2")
plt.legend()
plt.xticks(np.arange(0,len(month)),labels=month)
```

```python
plt.pie([22,34,56,78,92,111],labels=["A","B","C","D","E","F"],autopct="%1.1f%%",pctdistance=0.8,wedgeprops={"linewidth":1.2,"edgecolor":"black"})
plt.legend()
```

```python
plt.figure(figsize=(6,6))
north=[23,29,33,39,44,51,58,88,97,111,128,130,150]
east=[21,28,38,44,49,52,59,67,77,87,100,109,141]
west=np.random.randint(18,170,size=13)
south=np.random.randint(18,172,size=13)
sales=np.random.randint(22,200,size=13)
plt.subplot(2,2,1)
plt.plot(sales,north,label="north")
plt.legend()
plt.subplot(2,2,2)
plt.plot(sales,east,label="east")
plt.legend()
plt.subplot(2,2,3)
plt.plot(sales,west,label="west")
plt.legend()
plt.subplot(2,2,4)
plt.plot(sales,south,label="south")
plt.legend()
plt.show()
```

```python
fig,ax=plt.subplots(2,2,figsize=(3,3),sharex=True,sharey=True)
ax1=ax[0,0]
ax2=ax[0,1]
ax3=ax[1,0]
ax4=ax[1,1]
ax1.plot(np.random.randint(10,100,size=10),np.arange(12,50,4),label="First",color="green")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend()
ax2.plot(np.random.randint(10,100,size=10),np.arange(12,50,4),label="Second")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.legend()
ax3.plot(np.random.randint(10,100,size=10),np.arange(12,50,4),label="Third",color="gold")
ax3.set_xlabel("X")
ax3.set_ylabel("Y")
ax3.legend()
ax4.plot(np.random.randint(10,100,size=10),np.arange(12,50,4),label="Forth",color="pink")
ax4.set_xlabel("X")
ax4.set_ylabel("Y")
ax4.legend()
ax2.set_title("GRAPH")
plt.tight_layout()
```

```python
x=np.linspace(0,4*np.pi,20)
x1=np.sin(x)
x2=2*x1
plt.plot(x,x1,marker="*",color="pink")
plt.plot(x,x2,marker="^",color="black")
plt.fill_between(x,x1,x2,color="red")
plt.show()
```

```python
sbn.get_dataset_names()
iris=sbn.load_dataset("iris")
```

```python
iris.species.unique()
```

```python
iris.plot("sepal_width","sepal_length",kind="scatter",c="petal_length",colormap="cool",colorbar=True)
```

```python
sbn.scatterplot(x=iris.petal_length,y=iris.sepal_length,hue=iris.species)
```

```python
iris.plot(kind="kde",alpha=0.3)
```

```python
iris.plot(kind="hist",bins=60)
```

```python
pd.plotting.scatter_matrix(iris,color="gold",hist_kwds={"color":"red"})
plt.tight_layout()
```

```python
pd.plotting.parallel_coordinates(iris,class_column="species",cols=["sepal_length","petal_width"])
```

```python
sbn.displot(iris,x="sepal_length",kde=True)
```

```python
sbn.scatterplot(iris,x=iris.sepal_length,y=iris.sepal_width,hue="species",palette=["gold","b","r"])
sbn.rugplot(iris,x=iris.sepal_length,y=iris.sepal_width,hue="species")
```

```python
sbn.jointplot(iris,x=iris.sepal_length,y=iris.sepal_width,kind="reg")
```

```python
fig,ax1=plt.subplots(figsize=(3,3))
sbn.regplot(iris,x=iris.sepal_length,y=iris.sepal_width,ax=ax1)
sbn.scatterplot(iris,x=iris.sepal_length,y=iris.sepal_width,hue="species",palette=["gold","b","r"],ax=ax1)
```

```python
sbn.pairplot(iris,size=1.1,hue="species",hue_order=["versicolor","setosa","virginica"],kind="kde",diag_kind="hist")
```
