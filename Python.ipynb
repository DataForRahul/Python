```python
import pandas as pd
import numpy as np
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date


date1 = pd.Series(pd.date_range('2012-1-1 12:30:10', periods=7, freq='2M'))
date2 = pd.Series(pd.date_range('2013-3-11 14:33:45', periods=7, freq='W'))

df = pd.DataFrame(dict(Start_date = date1, End_date = date2))
print(df)
```

Series.dt can be used to access the values of the series as datetimelike and return several properties

```python
df["difference in seconds"]=(df["End_date"]-df["Start_date"]).dt.total_seconds()
df
```

```python
df["difference in minutes"]=(df["End_date"]-df["Start_date"]).dt.total_seconds()/60
df
```

```python
df["difference in hours"]=(df["End_date"]-df["Start_date"]).dt.total_seconds()/60/60
df
```

```python
df["weeks_new"]=df["End_date"]-df["Start_date"]
df["weeks_new"]=df["weeks_new"]/np.timedelta64(1,"W")  # here timedelta64(1,"W") 1 means 7 days or 1 week,if 2 then 14 days is 1 week
df
```

```python
df["weeks_new"]=df["End_date"]-df["Start_date"]
df["days_new"]=df["weeks_new"]/np.timedelta64(1,"D")
df
```

```python
df["Days_"]=df["End_date"]-df["Start_date"]
df
```
