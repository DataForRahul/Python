```python
class User:
    def __init__(self, pin, balance, name):
        self.balance = balance
        self.correct_pin = pin
        self.name = name

    def update_pin(self, new_pin):
        self.correct_pin = new_pin
        print("PIN successfully changed.")

    def update_balance(self, amount, transaction_type):
        if transaction_type == 'withdraw':
            if amount <= self.balance:
                self.balance -= amount
                print(f'Your available balance is {self.balance}')
            else:
                print('Insufficient Balance')
        elif transaction_type == 'deposit':
            self.balance += amount
            print(f'New Balance is {self.balance}')

def verify_pin(user):
    try:
        pin = int(input('Enter Your Pin: '))
    except ValueError:
        print("Invalid input! Please enter a numeric PIN.")
        return False

    if pin != user.correct_pin:
        print("Incorrect pin")
        return False
    return True

def atm_simulation(user):
    print(f'''
            ***** Welcome {user.name} To ATM Machine Simulator *****
         ''')

    if not verify_pin(user):
        print("Thank you")
        return

    while True:
        print('''
              Options you can exercise are:
              1) Balance
              2) Withdraw
              3) Deposit
              4) Change pin
              5) Exit
          ''')
        try:
            choice = int(input('Select Your Transaction from the above options: '))
        except ValueError:
            print("Invalid input! Please enter a number from the options.")
            continue

        if choice == 1:
            print(f'Available A/C Balance Is {user.balance}')
        elif choice == 2:
            try:
                withdraw = int(input('Enter Amount: '))
            except ValueError:
                print("Invalid input! Please enter a numeric amount.")
                continue

            user.update_balance(withdraw, 'withdraw')
        elif choice == 3:
            try:
                deposit = int(input('Enter Amount: '))
            except ValueError:
                print("Invalid input! Please enter a numeric amount.")
                continue

            user.update_balance(deposit, 'deposit')
        elif choice == 4:
            try:
                new_pin = int(input("Enter the new pin: "))
            except ValueError:
                print("Invalid input! Please enter a numeric PIN.")
                continue

            user.update_pin(new_pin)

            print("Please verify your new PIN to continue.")
            if not verify_pin(user):
                print("PIN verification failed. Returning to the main menu.")
                continue
        elif choice == 5:
            print("Thank you!")
            break
        else:
            print('Incorrect choice')
    return

def main():
    # Create a User instance for "Rahul"
    Rahul = User(pin=1998, balance=10000, name="Rahul")
    # Create another User instance for "Sonu"
    u1 = User(pin=1234, balance=20000, name="Sonu")

    # Uncomment the lines below to test the users one at a time
    # Run the ATM simulation for Rahul
    atm_simulation(Rahul)

    # Run the ATM simulation for u1
    # atm_simulation(u1)

if __name__ == "__main__":
    main()
```
